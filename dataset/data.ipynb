```python
import pandas as pd
import os
import glob 
```

```python
pathern = './sephora/*.csv'
all_files = glob.glob(os.path.join(pathern))
df_list = []
for file in all_files:
    df = pd.read_csv(file)
    df_list.append(df)
    print(f"{file} : {df.shape}")
```

```python
df_concat = pd.concat(df_list, axis=0, ignore_index=True)
df_concat["target_url"]="https://www.sephora.com/"+df_concat["target_url"]
```

```python
df_concat.to_csv('./sephora/sephora_full_dataset.csv', index=False)
```
